```python
import tensorflow as tf 
import numpy as np 
import tensorflow.keras as keras 
```

```python
def build_model():
    # Variable-length sequence of integers
    text_input_a = keras.Input(shape=(None,), dtype='int32')

    # Variable-length sequence of integers
    text_input_b = keras.Input(shape=(None,), dtype='int32')

    # Embedding for 1000 unique words mapped to 128-dimensional vectors
    shared_embedding = keras.layers.Embedding(1000, 128)

    # We reuse the same layer to encode both inputs
    encoded_input_a = shared_embedding(text_input_a)
    encoded_input_b = shared_embedding(text_input_b)

    # two logistic predictions at the end
    prediction_a = keras.layers.Dense(1, activation='sigmoid', name='prediction_a')(encoded_input_a)
    prediction_b = keras.layers.Dense(1, activation='sigmoid', name='prediction_b')(encoded_input_b)

    # this model has 2 input, and 2 outputs
    # in the middle we have a share model
    model = tf.keras.Model(inputs=[text_input_a, text_input_b], outputs=[prediction_a, prediction_b])

    keras.utils.plot_model(model, to_file='share_model.png')
```

```python
build_model()
```
